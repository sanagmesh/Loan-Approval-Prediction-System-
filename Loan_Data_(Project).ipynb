**4. Data Preprocessing**

```python
df1=df.copy(deep=True)
```

```python
df1.isna().sum()
```

Removing unnecessary columns

The following columns were removed because they were not useful for prediction:


1.   Married
2.   Education
3.   Self_Employed


```python
df1.drop(columns=['Married','Education','Self_Employed'], inplace=True)
df1.head()
```

Convert the object type columns into the integer type Data:

1.   Loan_ID
2.   Gender
3.   Property_Area
4.   Loan_Status



```python
df1['Loan_ID'] = df1['Loan_ID'].str.replace('LP', '').astype(int)
```

```python
df1['Gender'].value_counts()
```

```python
mapping={"Male":1, "Female":0}
df1["Gender"]=df1 ['Gender'].map(mapping)
```

```python
df1['Property_Area'].value_counts()
```

```python
mapping={"Semiurban":2, "Urban":1,"Rural":0}
df1 ['Property_Area']=df1 ['Property_Area'].map(mapping)
```

```python
df1['Loan_Status'].value_counts()
```

```python
mapping={"Y":1, "N":0}
df1 ['Loan_Status']=df1 ['Loan_Status'].map(mapping)
```

Handling missing values

Missing values in the following columns were filled using the mean value:


1.   Gender
2.   Dependents
3.   LoanAmount
4.   Loan_Amount_Term
5.   Credit_History



```python
df1.isna().sum()
```

```python
df1['Gender'].fillna(df1['Gender'].mode()[0], inplace=True)
df1.head(25)
```

```python
df1.info()
```

```python
df1['Dependents'].value_counts()
```

```python
df1['Dependents'].fillna(df1['Dependents'].mode()[0], inplace=True)
```

```python
df1.info()
```

```python
df1['Dependents'] = df1['Dependents'].replace('3+', 3).astype(int)
```

```python
df1['LoanAmount'].fillna(df1['LoanAmount'].median(), inplace=True)
df1['Loan_Amount_Term'].fillna(df1['Loan_Amount_Term'].median(), inplace=True)
```

```python
df1.isna().sum()
```

```python
df1['Credit_History'].fillna(df1['Credit_History'].mode()[0], inplace=True)
```

```python
df1.isna().sum()
```

```python
x=df1.drop(columns=['Loan_ID','Loan_Status'])
y=df1['Loan_Status']
```

**6.Model Training**

The dataset was split into training and testing sets.

Training data: 80%

Testing data: 20%

Two machine learning models were trained:

Logistic Regression

Random Forest Classifier

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=2)
```

**7.Model Evaluation**


The model was evaluated using **accuracy, confusion matrix, and classification report**.

Random Forest performed better than Logistic Regression and achieved about **74% accuracy**. The model predicts loan approval cases more accurately than rejection cases.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```

```python
model = LogisticRegression()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import gradio as gr
import pandas as pd

# Assuming 'rf' (RandomForestClassifier model) and 'x' (DataFrame of features)
# are already defined in the environment from previous executed cells.

def predict_loan_status_adjusted(gender, dependents_str, applicant_income, coapplicant_income,
                        loan_amount, loan_amount_term, credit_history, property_area):

    # Convert '3+' to integer 3 for 'Dependents'
    dependents = int(dependents_str.replace('+', ''))

    # Create a DataFrame from the inputs, ensuring column order matches x.columns
    input_data = pd.DataFrame([[gender, dependents, applicant_income, coapplicant_income,
                                loan_amount, loan_amount_term, credit_history, property_area]],
                              columns=x.columns)

    prediction = rf.predict(input_data)

    if prediction[0] == 1:
        return "Loan Approved ✅"
    else:
        return "Loan Denied ❌"

# Define Gradio inputs based on feature types and their mappings
inputs = [
    gr.Radio(choices=[("Male", 1), ("Female", 0)], label="Gender", value=1), # Default to Male
    gr.Radio(choices=["0", "1", "2", "3+"], label="Dependents", value="0"), # Default to 0 dependents
    gr.Number(label="Applicant Income", minimum=0, value=5000),
    gr.Number(label="Coapplicant Income", minimum=0.0, value=2000.0),
    gr.Number(label="Loan Amount (in thousands)", minimum=0.0, value=150.0),
    gr.Number(label="Loan Amount Term (in months)", minimum=0.0, value=360.0),
    gr.Radio(choices=[("Yes", 1.0), ("No", 0.0)], label="Credit History", value=1.0), # Default to Yes
    gr.Radio(choices=[("Semiurban", 2), ("Urban", 1), ("Rural", 0)], label="Property Area", value=2) # Default to Semiurban
]

# Define the output component
output = gr.Label()

# Define custom CSS for a visually appealing interface
css = """
body {
    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
    background-color: #e0f2f7; /* Light blue background */
}
.gradio-container {
    max-width: 900px;
    margin: 40px auto;
    border-radius: 20px;
    box-shadow: 0 15px 40px rgba(0, 0, 0, 0.15);
    background-color: #ffffff;
    border: 1px solid #b3e5fc;
}
.gr-button.primary {
    background-color: #0288d1; /* Deep Sky Blue */
    color: white;
    border-radius: 10px;
    padding: 12px 25px;
    font-size: 17px;
    font-weight: bold;
    transition: background-color 0.3s ease;
}
.gr-button.primary:hover {
    background-color: #0277bd; /* Slightly darker blue on hover */
}
.gr-label {
    font-size: 28px;
    font-weight: bold;
    color: #263238; /* Dark grey text */
    text-align: center;
    padding: 25px;
    border-top: 2px solid #e0f2f7;
    background-color: #f5f5f5;
    border-bottom-left-radius: 18px;
    border-bottom-right-radius: 18px;
}
h1 {
    color: #01579b; /* Darker blue for main title */
    text-align: center;
    font-size: 38px;
    margin-bottom: 15px;
    text-shadow: 1px 1px 2px rgba(0,0,0,0.05);
}
h3 {
    color: #455a64; /* Medium grey for description */
    text-align: center;
    font-size: 19px;
    margin-bottom: 35px;
}
.gr-radio-group {
    font-weight: 500;
    color: #37474f;
}
"""

# Create the Gradio interface
iface = gr.Interface(
    fn=predict_loan_status_adjusted,
    inputs=inputs,
    outputs=output,
    title="<span style='color:#01579b;'>🏦 Loan Approval Predictor 📈</span>", # Title with color
    description="<h3>Input the applicant's details below to get an instant prediction on their loan approval status.</h3>", # Description with styling
    css=css,
    theme=gr.themes.Soft() # Use the 'Soft' theme for a modern look
)

# Launch the interface
iface.launch(share=True)
```
